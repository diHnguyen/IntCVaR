```julia
using JuMP 
using Gurobi
using LightGraphs
using DataFrames
using CSV
using TimerOutputs
using Dates
using Random

gurobi_env = Gurobi.Env()
setparams!(gurobi_env, Heuristics=0.0, Cuts = 0, OutputFlag = 0)

α = 0.3
p = [1]
m = Model(() -> Gurobi.Optimizer(gurobi_env)) # If we want to add # in Gurobi, then we have to turn of Gurobi's own Cuts 
g = [2]
@variable(m, w)
@variable(m, 1e6 >= z[1:500000] >= 0)

# @constraint(m,con[1], w <= SP_init + sum(yy[i]*x[i]*d[i] for i=1:Len) + z[1]  )
# @constraint(m,con[2], w >=0  )
#     println(sum(yy[i]*x[i]*d[i] for i=1:Len))
@objective(m, Max, α*w - sum(p[i]*z[i] for i = 1:length(p)) )#w - sum(s[k] for k=1:length(s))/length(s) )


start = time()
for iter = 1:18
    println("Iter ", iter)
    K = 2^iter
    p = ones(K)./K
    for i = 1:K
        rhs = rand(1:100)
        if i > length(g)
            push!(g, rhs)
        elseif rhs < g[i]
            g[i] = rhs
        end
        sort(g)
        
        @constraint(m, w <= rhs + z[i]  )
    end
    temp_p = 0
    count = K
    while temp_p + 1/K < α
        p[count] = 0
        count = count-1
        temp_p = temp_p + 1/K
    end
    @objective(m, Max, α*w - sum(p[i]*z[i] for i = 1:K) )
    optimize!(m)
    println("MP = ", JuMP.objective_value.(m))
    println("K = ", K)
end

elapsed = time() - start
println(elapsed)
    
```
