```julia

using Plots
K = 10^6

# r = [3,4]
# a = [0,1]
function plotPDF(K, r, a)
    dim = length(r)
    c = Array{Float64}(undef, K,dim)
    # c[:,1] = rand(K)*r1
    # c[:,2]= rand(K)*r2 .+ a2
    for m =1:length(r)
       c[:,m] = rand(K).*r[m].+a[m]
    end
    # println("1")
    C = sum(c[:,m] for m = 1:length(r))
    C = sort!(C)
    println("C = ",C[1:10])
    println("max = ", maximum(C))
    println("min = ", minimum(C))
    # println("lengthC ", length(C))
    #
    println("r =", r)
    println("a = ", a)
    d = sum(r[m] for m=1:length(r))# r2+r1
    increment = 0.1
    N = Int64(d/increment)
    # println("N = ",N)
    s = collect(0:N)
    s = s.*d./N .+ sum(a[m] for m=1:length(r))
    # println("2")
    # println("s = ", s)

    p = zeros(length(s)-1)
    # println("length(p) =", length(p))
    pointer = 1
    i = 1
    while i < K
        g = C[i]
    #     println("(",s[pointer],", ", s[pointer+1],")")
        if g >= s[pointer] && g <= s[pointer+1]
    #         println("true")
            p[pointer] = p[pointer] + 1
        else
    #         println(pointer)
    #         println("i = ", i)
            pointer = pointer+1
        end
        i = i+1
    end
    p=p./K
    #     println(p)
    # for pointer = 1:60
    # #     println("(",s[pointer],", ", s[pointer+1],"): ", p[pointer])
    # #     println("(1,", 1 + i*(6/20),")")

    # end
    # using Pkg
    # Pkg.add("Plots")
    # println("length(p) = ",length(p))
    
    # println("C1 = (0,1)")
    # println("C2 = (1,5)")
    
    return C,s,p
end


function plotSPcost(K, r, a)
    dim = length(r)
    c = Array{Float64}(undef, K,dim)
    # c[:,1] = rand(K)*r1
    # c[:,2]= rand(K)*r2 .+ a2
    for m =1:length(r)
       c[:,m] = rand(K).*r[m].+a[m]
    end
    # println("1")
    C = sum(c[:,m] for m = 1:length(r))
    C = sort!(C)
#     println("C = ",C[1:10])
#     println("max = ", maximum(C))
#     println("min = ", minimum(C))
    # println("lengthC ", length(C))
    #
#     println("r =", r)
#     println("a = ", a)
    
#     d = sum(r[m] for m=1:length(r))# r2+r1
    
    increment = 0.1
    N = Int64(1/increment)
    # println("N = ",N)
    s = collect(0:N)
    s = s./N
#     s = s.*d./N .+ sum(a[m] for m=1:length(r))
    # println("2")
    println("s = ", s)

    p = zeros(length(s)-1)
    # println("length(p) =", length(p))
    pointer = 1
    
    g_sum = zeros(length(s)-1)
    
    for i = 1:K
        if K/i > pointer
            pointer = pointer + 1
        end
        g = C[i]
    #     println("(",s[pointer],", ", s[pointer+1],")")
        if g >= s[pointer]* && g <= s[pointer+1]
            p[pointer] = p[pointer] + 1
            g_sum[pointer] = g_sum[pointer] + g
        else
            pointer = pointer+1
        end
        
        i = i+1
    end
    
    g_final = zeros(length(s)-1)
    p_final = zeros(length(s)-1)
    for i = 1 : length(p)
        g_final[i] = sum(g_sum[j] for j = 1:i)
        p_final[i] = sum(p[j] for j = 1:i)
    end
#     println("pfinal ")
    p=p./K
    #     println(p)
    # for pointer = 1:60
    # #     println("(",s[pointer],", ", s[pointer+1],"): ", p[pointer])
    # #     println("(1,", 1 + i*(6/20),")")

    # end
    # using Pkg
    # Pkg.add("Plots")
    # println("length(p) = ",length(p))
    
    # println("C1 = (0,1)")
    # println("C2 = (1,5)")
    
    return g_final, p_final
end

K = 10^6

cL = [1.1,2.3,3.5, 5.7]
cU = [3.2,6.4,9.6, 20.8]

r = cU-cL#[2,4,3]
a = cL #[1,2,4]
C,s,p = plotPDF(K,r,a)
g_final,p_final = plotSPcost(K, r, a)
# plot(s[1:length(s)-1],p)
plot(g_final, p_final)
```
